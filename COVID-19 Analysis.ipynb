## Trend Analysis

```python
x = data.index
y = data["Total_cases"]
sns.scatterplot(x,y)
plt.xticks(rotation=90)
plt.xlabel('Date', fontsize=10)
plt.ylabel('Total cases', fontsize=10)
```

```python
size = (16,10)
fig, ax = plt.subplots(figsize = size)
ax = plt.plot(data.index,data['Total_cases'])
plt.xticks(rotation=90)
plt.xlabel('Date', fontsize=10)
plt.ylabel('Total cases', fontsize=10)
```

```python
ax = sns.boxplot(x=data.index,y='Total_cases',data=data)

ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right")
plt.tight_layout()
plt.show()
```

```python
ax = sns.barplot(x=data.index, y='Total_cases', data=data)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right")
plt.tight_layout()
plt.show()
```
